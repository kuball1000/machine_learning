```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)

print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba próbek: {m}')
```

```python
X1 = X1.reshape(m, 1)
Y = Y.reshape(-1, 1)
print(X1)
print(X1.shape)
```

```python
bias = np.ones((m, 1))
print(bias)
print(bias.shape)
```

```python
X = np.append(bias, X1, axis=1)
print(X)
print(X.shape)
```

```python
eta = 0.01
weights = np.random.randn(2,1)
print(X)
print(weights)
```

```python
intercept = []
coef = []

for i in range(3000):
  gradient = (2/m) * X.T.dot(X.dot(weights) - Y)
  weights = weights - eta * gradient
  intercept.append(weights[0][0])
  coef.append(weights[1][0])

print(weights)
```

```python
df = pd.DataFrame(data={'intercept': intercept, 'coef': coef})
df.head()
```

```python
px.line(df, y='intercept', width=1600, title='Dopasowanie: intercept')
```

```python
px.line(df, y='coef', width=1600, title='Dopasowanie: coef')
```
